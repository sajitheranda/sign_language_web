```python
video_path = "temp/tmp7njta228"
save_path = "temp/new.json"

 # Initialize Wholebody model
wholebody = Wholebody() 

video_keypoints = []

# Sort and filter for only files matching the "XXX.jpg" format
images = sorted([img for img in os.listdir(video_path) if img.endswith(".jpg")])

for i in range(len(images)):
    image = images[i]

    # Extract the frame number and ensure it matches the index
    frame_id = int(image.split('.')[0].lstrip('0'))  # Convert "001" -> 1, "002" -> 2, etc.
    assert i + 1 == frame_id, f"Index {i + 1} is not equal to frame_id {frame_id} in {image}"
    
    image_path = os.path.join(video_path, image)
    print(image_path)
    img = cv2.imread(image_path)
    keypoints, scores = wholebody(img)

    video_keypoints.append(keypoints.tolist())

# Save keypoints as JSON
with open(save_path, 'w') as json_file:
    json.dump(video_keypoints, json_file)
```

```python
img_show = draw_skeleton(img, keypoints, scores, kpt_thr=0.5)


cv2.imshow('img', img_show)
cv2.waitKey()
```

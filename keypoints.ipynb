## Keypoint Generation

```python
import os
from pathlib import Path
import json
import cv2
from rtmlib import Wholebody, draw_skeleton
```

```python
pose = "models/Keypoints_models/rtmw-x_simcc-cocktail13_pt-ucoco_270e-384x288-0949e3a9_20230925/20230928/rtmpose_onnx/rtmw-x_simcc-cocktail13_pt-ucoco_270e-384x288-0949e3a9_20230925/end2end.onnx"
pose_input_size = (288, 384)
det = "models/Keypoints_models/yolox_x_8xb8-300e_humanart-a39d44ed/20230928/yolox_onnx/yolox_x_8xb8-300e_humanart-a39d44ed/end2end.onnx"
det_input_size = (640, 640)

device = 'cpu'  # cpu, cuda, mps
backend = 'onnxruntime'  # opencv, onnxruntime, openvino
openpose_skeleton = False  # True for openpose-style, False for mmpose-style


# wholebody = Wholebody() 
wholebody = Wholebody(to_openpose=openpose_skeleton,
                    det = det ,
                    det_input_size = det_input_size,
                    pose = pose,
                    pose_input_size = pose_input_size,
                    mode='balanced',  # 'performance', 'lightweight', 'balanced'. Default: 'balanced'
                    backend=backend, device=device)
```

```python
video_path = "temp/tmp_2zp6rht"
save_path = "temp/new.json"

 # Initialize Wholebody model
# wholebody = Wholebody() 

video_keypoints = []

# Sort and filter for only files matching the "XXX.jpg" format
images = sorted([img for img in os.listdir(video_path) if img.endswith(".jpg")])

for i in range(len(images)):
    image = images[i]

    # Extract the frame number and ensure it matches the index
    frame_id = int(image.split('.')[0].lstrip('0'))  # Convert "001" -> 1, "002" -> 2, etc.
    assert i + 1 == frame_id, f"Index {i + 1} is not equal to frame_id {frame_id} in {image}"
    
    image_path = os.path.join(video_path, image)
    print(image_path)
    img = cv2.imread(image_path)
    keypoints, scores = wholebody(img)

    video_keypoints.append(keypoints.tolist())

# Save keypoints as JSON
with open(save_path, 'w') as json_file:
    json.dump(video_keypoints, json_file)
```

```python
img_show = draw_skeleton(img, keypoints, scores, kpt_thr=0.5)


cv2.imshow('img', img_show)
cv2.waitKey()
```
